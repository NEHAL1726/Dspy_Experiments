# Credit Card Fraud Detection
## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("creditcard.csv")
```

## Graphical Analysis

### 1. Univariate Analysis

What is being analyzed:
Distribution of transaction amounts

Type:
Univariate Graphical Analysis

```python
sns.histplot(df['Amount'], bins=50)
plt.title("Distribution of Transaction Amount")
plt.show()
```

Observation:
Most transactions are of small amounts.

Conclusion:
Fraud detection cannot depend only on transaction amount.

### 2. Multivariate Analysis

What is being analyzed:
Relationship between transaction amount and fraud

Type:
Multivariate Graphical Analysis

```python
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Amount vs Fraud")
plt.show()
```

Observation:
Fraud transactions vary widely in amount.

Conclusion:
Transaction amount alone is not sufficient to detect fraud.

## Non-Graphical Analysis

### 3. Class Distribution

What is being analyzed:
Proportion of fraud vs normal transactions

Type:
Univariate Non-Graphical Analysis

```python
df['Class'].value_counts(normalize=True)
```

Observation:
Fraud cases are extremely low compared to normal cases.

Conclusion:
The dataset is highly imbalanced.

### 4. Correlation Analysis

What is being analyzed:
Relationship between features and fraud

Type:
Multivariate Non-Graphical Analysis

```python
df.corr()['Class'].sort_values(ascending=False)
```

Observation:
Some features show correlation with fraud.

Conclusion:
These features are useful for building a prediction model.
