**Step 1: Library aur Data Load**

```python
# Install library
!pip install datasets

# Import libraries
import pandas as pd
import re
from datasets import load_dataset

# Load the data (Ye step bahut zaruri hai, isi se 'df' banega)
print("Loading dataset...")
dataset = load_dataset("imdb")
df = pd.DataFrame(dataset['train'].select(range(2000)))
print("Dataset loaded successfully!")

# Check for Nulls
print("\nMissing values count:\n", df.isnull().sum())

# Dropping duplicates
initial_shape = df.shape
df = df.drop_duplicates()
print(f"\nDuplicates removed. Shape changed from {initial_shape} to {df.shape}")

# Text Cleaning Function
def clean_text(text):
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text.lower()

# Applying the cleaning
df['text'] = df['text'].apply(clean_text)
print("\nText cleaning complete.")
print(df.head())

# Save to CSV
df.to_csv('cleaned_imdb_data.csv', index=False)
print("\nFile saved as 'cleaned_imdb_data.csv'")
```
