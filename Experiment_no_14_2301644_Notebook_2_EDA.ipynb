# **Notebook 2: Exploratory Data Analysis (EDA)**
**Project:** Sentiment Analysis of Movie Reviews

**Objective:** To explore the dataset through statistical and graphical methods to understand review patterns and word distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Load the cleaned dataset from Notebook 1
# Ensure 'cleaned_imdb_data.csv' is uploaded to your Colab session
df = pd.read_csv('cleaned_imdb_data.csv')

# Adding a helper column to calculate the length of each review
df['review_length'] = df['text'].apply(len)

print("Data successfully loaded. Here is a preview:")
print(df.head())
```

# 1. Non-Graphical Analysis (Descriptive Statistics)
Objective: To understand the basic structure and numerical distribution of our dataset.

```python
# General information about the dataset
print("--- Dataset Info ---")
print(df.info())

# Statistical summary of review lengths
print("\n--- Review Length Statistics ---")
print(df['review_length'].describe())

# Check the count of each sentiment label
print("\n--- Sentiment Label Count ---")
print(df['label'].value_counts())
```

# Observation:

The dataset consists of 1,997 reviews.

The average review length is approximately 1,300 characters.

Currently, the dataset is dominated by negative reviews (label 0), which serves as our primary focus for this analysis.

# 2. Graphical Analysis: Distribution of Review Lengths
Objective: To visualize how long the movie reviews typically are.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['review_length'], bins=50, color='royalblue', kde=True)
plt.title('Distribution of Review Lengths')
plt.xlabel('Number of Characters')
plt.ylabel('Frequency (Number of Reviews)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Observation:**

 The distribution is "right-skewed." Most reviews are between 500 and 1,500 characters long. There are a few very long reviews that exceed 4,000 characters, which might contain more detailed critiques.

# 3. Graphical Analysis: Top 20 Most Frequent Words
Objective: To identify the most common words used in movie reviews.

```python
# Flatten all text into a single list of words
all_words = ' '.join(df['text']).split()
word_freq = Counter(all_words)

# Convert to a DataFrame for plotting
top_words_df = pd.DataFrame(word_freq.most_common(20), columns=['Word', 'Count'])

# Create a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(data=top_words_df, x='Count', y='Word', palette='viridis')
plt.title('Top 20 Most Frequent Words in the Dataset')
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.show()
```

**Observation:**

The most frequent words are "movie," "film," "bad," and "even." This confirms that our cleaning was successful and that the dataset is indeed focused on cinematic reviews.

# Final Conclusion of EDA
Data Consistency: The text data is uniform and ready for vectorization.

Text Complexity: Review lengths are varied, so our model must be robust enough to handle different input sizes.

Sentiment Clues: High frequency of negative-leaning words like "bad" and "trash" (visible in common words) indicates strong sentiment features for training.
